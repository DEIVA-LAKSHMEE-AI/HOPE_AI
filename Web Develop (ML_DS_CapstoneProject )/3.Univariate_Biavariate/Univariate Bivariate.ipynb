```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\AI course\ML_DS_CapstoneProject\Dataset\Placement_Data_Cleaned.csv")
print("Shape:", df.shape)
df.head()
```

```python
# ==========================================
# Univariate Analysis - Target Distribution
# ==========================================
print(df['status'].value_counts())
print(df['status'].value_counts(normalize=True))

```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='status', data=df)
plt.title("Placement Status Distribution (1=Placed, 0=Not Placed)")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
# ==========================================
# Univariate Analysis - Numeric Feature Distributions
# ==========================================
numeric_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']

df[numeric_cols].hist(figsize=(12, 8), bins=15)
plt.suptitle("Distribution of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# Bivariate Analysis - Numeric Features vs Target
# ==========================================
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='status', y=col, data=df)
    plt.title(f"{col} vs Placement Status")
    plt.xlabel("Status (0 = Not Placed, 1 = Placed)")
    plt.ylabel(col)
    plt.show()
```

```python
# ==========================================
# Correlation Heatmap
# ==========================================
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# Correlation with Target (sorted)
# ==========================================
print(df.corr(numeric_only=True)['status'].sort_values(ascending=False))
```
