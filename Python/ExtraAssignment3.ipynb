```python
# Catagory the people by their age like children, adult, citizen, senior citizen... 
num = int(input("age: ")) 
if (num <= 18): 
 print("Children") 
elif (num <= 35): 
 print("Adult") 
elif (num <= 60): 
 print("Citizen") 
else: 
 print("senior citizen") 
```
