```python
import pandas as pd
df = pd.read_csv(r"D:\AI course\ML_DS_CapstoneProject\Dataset\Placement_Data_Full_Class.csv")
```

```python
# ------------------------------------------
# Drop columns not usable for prediction
# ------------------------------------------
df = df.drop(columns=["sl_no", "salary"])
```

```python
# ------------------------------------------
# Encode the target variable
# ------------------------------------------
df["status"] = df["status"].map({"Placed": 1, "Not Placed": 0})
```

```python
# ------------------------------------------
# Encode categorical features
# ------------------------------------------
categorical_cols = ["gender", "ssc_b", "hsc_b", "hsc_s", "degree_t", "workex", "specialisation"]
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

```

```python
print("Shape after preprocessing:", df_encoded.shape)
print("\nColumns after encoding:")
print(df_encoded.columns.tolist())
print("\nPreview:")
print(df_encoded.head())
```

```python
# ==========================================
# Save Cleaned Dataset (feeds into next stage)
# ==========================================
df_encoded.to_csv(r"D:\AI course\ML_DS_CapstoneProject\Dataset\Placement_Data_Cleaned.csv", index=False)
print("Cleaned dataset saved successfully.")
```
